```python
import random
```

```python
name=input("enter your name:")

print("Goodluckk!!", name)

words=['rainbow', 'butterfly', 'mission mangal', 'cpp', 'python', '3 idiots',
       'dog', 'water', 'java', 'javascript', 'reverse', 'html', 'css',
       'taare zameen par', 'coding', 'maths', 'sceince','programming','stranger things',
       'red', 'blue', 'green', 'purple']

word= random.choice(words)

print("Guess the characters")

guesses=''
turns=10

while turns>0:
    failed=0

    for char in word:
        if char in guesses:
            print(char, end=" ")

        else:
            print("_")
            failed+=1


    if failed==0:
        print("Yayayayayya!!!! You win")
        print("The word is :", word)
        break

    print()

    guess=input("guess a character:")

    guesses+=guess

    if guess not in word:

        turns-=1
        print("OHH noooo!! Try againnn...")
        print("You Have", + turns, 'more guesses')

        if turns==0:
            print("You Loosee, Better Luck next time!!")
```
